# Емпирични функции на разпределение и свързани с нея статистики

Основна задача в мат. статистика е да се да се определи разпределението на наблюдаваната сл.в. Решаването на тази задача минава през няколко етапа, първият от които е графично представяне на функцията на разпределение и вида на съответната плътност на разпределение. Нека $X_1, \ldots, X_n$ са наблюдения над сл.в. $X$. Ще предполагаме, че наблюденията представляват проста сл. извадка от стойностите на $X$ и, че $X$ има функция на разпределение $F(x)$.

## Вариационен ред

Наредените във възходящ ред наблюдения от редицата се наричат **вариационен ред** на наблюденията. Величината $X_{(k)}$ се нарича $k$-тата поредна статистика.

## Емпирична функция на разпределение

Вариационният ред е първична форма на записване на стат. материал и може да се обработва по различни начини. Един подход е построяването на емпиричната (стат.) функция на разпределение на сл.в. $X$.

Функцията, определена с равенствата:

$$
F_n(x)= 
\begin{cases}
0 \quad \text{при} \quad x \leq X_{(1)} \\
\frac{k}{n} \quad \text{при} \quad X_{(k)} < x \leq X_{(k + 1)} \\
1 \quad \text{при} \quad x > X_{(n)}
\end{cases}
$$

се нарича **емпирична функция на разпределение**. От неравенствата става ясно, че $F_n(x)$ се получава, като на всяка от $n$-те наблюдавани ст-сти на $X$ се приписва вероятност $\frac{1}{n}$, т.е. $X$ е равномерно разпределена върху числата $X_1, \ldots, X_n$.

Емпиричната функция на разпределение е монотонна, непрекъсната отляво и има точки на прекъсване само при ст-ти на аргумента, съвпадащи с членовете на вар. ред. В точките на прекъсване големината на скока е кратна на $\frac{1}{n}$

## Числови х-ки на емпиричното разпределение

Средно аритметично: $\overline{\xi_n} = \int_{-\infty}^{+\infty} = xdF_n(x)$

Извадкова дисперсия $\sigma_{n}^2 = \frac{1}{n}\sum_{i=1}^n(\xi_i - \overline{\xi_n})^2$

С увеличаване на броя на набл. всички средни стат. х-ки клонят по вероятност към съответните числови х-ки на сл.в. $\xi$, която се наблюдава.

## Теорема на Гливенко - Кантели

Ако $F(x)$ е истинската ф-я на разпр. на сл.в. $\xi$, а $F_n(x)$ - емпиричната й функция, получена от $n$ независими наблюдения, то

$$P\big\{\sup_{-\infty < x < \infty} \big| F_n(x) - F(x) \big| \xrightarrow[n \rightarrow \infty] \, 0\big\} = 1$$

## Критерий на Колмогоров

Ако функцията на разп. $F(x)$ на сл.в. $\xi$ е непрекъсната, то при $n \rightarrow \infty$ е изпълнено

$$\lim_{n \rightarrow \infty}P\{\sqrt{n}D_n < z\} = K(z) = 1 - 2 \sum_{k=1}^{\infty}(-1)^{k-1}e^{-2k^2z^2}$$

## $\chi^2$ критерий за съгласие

Нека $\xi_1, \ldots, \xi_n$ е проста извадка от набл. от сл.в. $\xi$. Искаме да разберем дали предположението ни $\xi$ да има конкретен закон на разпределение е разумно.

Нека имаме $H_0$ - опитните данни са ст-ти на сл.в. $\xi$, имаща ф. на разпр. $F(x)$. Трябва да изберем метод за проверка на това, дали опитните данни са съгласувани с $H_0$. Един от най-често използваните критерий за проверка на тези условия е $\chi^2$ критерият на Пирсън.

Разделяме цялата област на изменение на сл.в. $\xi$ на $l$ интервала $\Delta_1, \ldots, \Delta_l$ и пресмятаме броя $\hat{m}, \; j = 1, \ldots, l$ на величините $\xi_i, \; i=1, \ldots, n$, принадлежащи на $\Delta_j$. Може да определим вероятността $p_j$ сл. в. $\xi$ да попадне в интервала $\Delta_j$:

$$p_j = P\{\xi_i \in \Delta_j\}, \quad j= 1, \ldots, l.$$

Тогава теор. брой на ст-тите на сл.в. $\xi$, попаднали в интервала $\Delta_j$, може да се пресметне по формулата $np_j = m_j$.

Ако емпиричните честоти силно се отличават от теоретичните, то $H_0$ се отхвърля, в противе сл. се приема.

Ако $H_0$ е вярна, то като мярка за близост между $\hat{m_j}$  и $np_j$ за $j = 1, \ldots, l$ се използва критерият

$$\hat{\chi^2} = \sum_{j=1}^l\frac{(\hat{m_j} - np_j)^2}{np_j}$$

При вярна $H_0$ величината $\hat{\chi^2}$ има асимптотично разпределение както $\chi^2(k)$ с $k = l - r -1$ степени на свобода, където $r$ е бр. параметри на $F(x)$, пресметнати с помощта на извадката.

Критерият $\chi^2$ е незаменим в случаите, когато директно получаваме емпиричните стойности $\hat{m_j}, j = 1, 2, \ldots, l$, без да знаем точните наблюдения над сл.в. $\xi$.
