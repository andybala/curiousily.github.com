# Каноничен анализ 

## Каноничен Кореспондентен Анализ

Анализът на главните компоненти Principal Component Analysis (PCA) ни предоставя начин за намаляне на размерността на данните, които имаме. Това ни позволява да ги визуализираме в едномерно, двумерно или тримерно пространство. PCA работи само с количествени данни.

Подобно на PCA, Кореспондетният анализ (КА) използва нивата на променливите за намиране на "нови" променливи и основните (координатните) оси, които най-добре обясняват наблюденията. Този подход ни позволява да работим с категорийни данни.

### Кореспондентен анализ

Нека имаме променлива, която разбива популацията на краен брой подпопулации, т.е. категорийна променлива. Нейните възможни ст-сти се наричат нива или категории. Данни, съдържащи наблюдения над такава променлива се наричат категорийни данни. Те могат да се измерват в номинална или категорийна скала. Категорийните данни се представят обикновено в честотни таблици. Множеството от относителните честоти на една категорийна променлива се нарича профил.

КА анализира разликите между относителните честоти на честотните таблици. За намиране на различието между категориите се използва $\chi^2$ разстоянието.

#### Инертност

Интертността измерва разпръстнатостта на категорийните данни. Дефинира се като:

$$\frac{\chi^2}{n}$$

където $n$ е обема на извадката. Интертността се изменя в интервала $[0; 2]$. Тя е близка до $0$, когато профилите лежат близо до средния профил (средата на новата координатна система).

#### Пасивни точки

Когато имаме допълнителна информация за редовете или стълбовете на честотната таблица, може да я визуализираме чрез допълни точки, които наричаме **пасивни**. Те не влияят на решението на КА, само допринасят за получаване на нова информация от данните. Когато искаме полученото решение да е пряко свързано с тези допълнителни точки, те трябва да бъдат представени като активни.

### Каноничен Кореспондентен Анализ

КА намира основните (координатните) оси на пространство с по-ниска размерност като се стреми да запази разстоянието между точките. Каноничният кореспондентен анализ (Canonical Correspondence Analysis, ККА) е едно решение на КА, което включва външни обясняващи променливи, които често се наричат променливи на "околната среда". Те могат да бъдат измерени в интервалната скала. Между осите на КА и външните променливи има линейна зависимост.

ККА ограничава търсенето на оптималните основни оси до част от пространството, наречено **канонично пространство**. В каноничното пространство КА намира най-добрата размерност за визуализиране.

Резултатът от ККА в каноничното пространство съдържа координатите на редовете и стълбовете от честотната талибца, както в КА.

#### Категорийни променливи в ККА

За добавяне на категорийни променливи в ККА се използват фиктивни (dummy) променливи. В графиката на ККА към тези фиктивни променливи не сочат стрелки, както се визуализират непрекъснатите променливи. Те се визуализират само с текст. Всяко ниво на категорийната променлива се представя в графиката на ККА.

## Каноничен Корелационен Анализ (ККОА)

Каноничният Корелационен Анализ разглежда връзката между две множества от случайни величини. Той е разширения на корелационния анализ.

Каноничната корелация (КК) намира корелацията между претегленото средно от първото множество $X$ и това от второто множество $Y$. Теглата се конструират, така че да максимизират корелацията между средните. Тази корелация се нарича "първа канонична корелация".

Може да създадем друго множество от претеглени средни, несвързани с първи, и изчислим тяхната корелация. Този процес продължава, докато броят на каноничните корелации не стане равен на този от броя променливи в множество $X$ или $Y$.

### Употреба

Каноничният Корелационен Анализ основно се употребява за:

- Намаляване количеството данни: обяснява ковариацията между $X$ и $Y$ посредством малко на брой линейни комбинации. Тези комбинации наричаме канонични променливи.
- Интерпретация на данните: намира променливите, които са важни за обясняването на ковариацията между $X$ и $Y$

### Изисквания

- Предполага се линейна зависимост между променливите. 
- Не се изисква данните да са нормално разпределени. Екстремни стойности (outliers) могат да доведат до получаване на неточни резултати. Входните данни трябва да бъдат предварително обработени.

### Корелационна матрица

Резултатът от ККОА е корелационната матрица $C$ за всички променливи. Тя може да се раздели на 4 подматрици:

- $R_{xx}$ - корелации между променливите в $X$
- $R_{yy}$ - корелации между променливите в $Y$
- $R_{xy}$ - корелации между $X$ и $Y$
- $R_{yx}$ - корелации между $Y$ и $X$

Тогава ККОА може да се дефинира като:

$$C = R_{yy}^{-1}R_{yx}R_{xx}^{-1}R_{xy}$$
