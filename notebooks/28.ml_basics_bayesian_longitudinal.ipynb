#  Бейсов анализ при повторени във времето изследвания.

Често се налага нуждата за оценка на даден подход/медикамент в медицината върху пациенти. Дизайнът на подобен вид изследвания се състои от предварително измерване - измерва се базовото състояние на пациента, назначаване на лечение и едно или повече измервания след интервенцията. Тук наблюдаваната промяна се асоциира с назначеното лечение. За оценка на ефективността на предложено лечение използваме повторени във времето изследвания.

## Повторени във времето изследвания (Longitudinal Data Analysis (LDA))

Изследвания в които се наблюдават измервания на дадена х-ка на всеки индивид, повторени във времето се наричат **повторени във времето изследвания**.

### Къде се използва?

- Клинични изследвания, които наблюдават хора или животни за определено време
- Изследвания за увеличаване или спад на растежа в агрикултурата, биологията и други

### Предимства на LDA

- Измерване на индивидуална промяна - предоставят възможност за наблюдаване на индивидуални модели (pattern)
- Отделяне на влиянието на времето - когато наблюдаваме промяна във времето е възможно да се използват много времеви скали. LDA може едновременно да характеризира няколко такива скали
- Контрол за групови ефекти - наблюдаваната група е фиксирана и промените през времето не са объркват от разликите в групата.

Нека х-ките за даден индивид са измерени многократно. Тези измервания ще отбележим с $y_{ij}$ където $i$ показва номер на индивид, а $j$ време на измерването, където $i = 1, \ldots, n$ и $j = 1, \ldots, m_i$ времеви периода. Нека всички измервания са направени в предварително планирани времеви периоди $t_{ij}$ и всички наблюдения са извършени в същите времеви периоди.

Редицата $\{y_{ij}\}_j = 1, \ldots, m_i$ образува времеви ред от измерванията за всеки индивид ( $S_i$, защото може да имаме различен брой наблюдения във времето, т.е. да имаме липсващи данни). Нека параметърът за средна ст-ст за всяко наблюдение е $\mu_{ij}$. Ще моделираме $\mu_{ij}$ като функция от зависими и независими от времето ковариати. Моделиране на ковариантната матрица на предсказваната променлива, $V_i$, също е важно за този вид изследвания. Подходяща структура може да направи изследването по-силно.

## Модели

### Модели с фиксирани ефекти

В подобен вид изследвания, най-често използваните модели са модели с миксирани ефекти:

$$Y_i = X_i\alpha + \epsilon_i$$

където $X_i$ е $S_ixp$ матрица с фиксирани стойности, $\alpha$ е $px_1$ вектор на параметрите, $\epsilon_i$ са грешките (сл.в.) - нормално разпределени със средна стойност вектора $O$ и ковариационна матрица $\Sigma$.

Математическото очакване на $Y_i$ е $EY_i = X_i\alpha$.
Ковариационната матрица на $Y_i$ е $\Sigma$, т.е. съвпада с ковариационната матрица на грешките $\epsilon_i$.

Решението на този модел може да се даде от традиционната (честотна) или Бейсовата статистика. Първият подход ни дава точкови оценки на параметрите, а вторият апостериорни вероятностни разпределения на параметрите на модела. 

Често, индивите проявяват изменчивост, която не може да бъде обяснена с този прост модел. Тогава се налага използването на модел със случайни ефекти:

### Модели със случайни ефекти

$$Y_i=X_i\alpha+W_i \beta_i+\epsilon_i$$

$W_i$ е $S_ixq$ е матрица от променливи, които не са сл. в. $q \leq p$, $\beta_i$ е $qx_i$ вектор от сл. ефекти, т.е. сл.в. които приемаме за нормално разпределени, със среден вектор $O$ и ковариационна матрица $V$. $\epsilon_i$ са грешките - те не са независими, защото имаме измервания върху едни и същи индивиди. Грешки са условно независими - при фиксирани $\beta_i$ те са независими.

За този модел ковариационната матрица на $Y_i$ е функция от $V$ на $\beta_i$ и $\Sigma$ на $\epsilon_i$. Тук $W_i$ и $X_i$ приемат фиксирани ст-сти, $\alpha$ е параметър, $\beta_i$ и $\epsilon_i$ са сл.в.

## Бейсов подход

За използваме описаните модели в Бейсовата рамка трябва да дефинираме априорни разпределения за сл.в. - параметъра $\alpha$, дисперсията $\sigma^2$ и ковариационната матрица $V$. Бейсовият подход разрешева параметрите да са сл.в., т.е. да имат вероятностни разпределения, които от своя страна също може да имат параметри със свое вероятностно разпределение и т.н. По този начин възниква йерархична дефиниция на моделите.

## Заключение

Моделите за анализ на повторени във времето изследвания отразяват връзката между наблюденията на един и същи индивид. Вероятностното разпределение на различните индивиди принадлежат към едно семейство (често то е Гаусово), но параметрите на сл. ефекти се различават.
