#  Корелационен анализ

Корелационният анализ (КА) е метод за измерване силата на корелационна зависимост между две или повече сл.в. Това се осъществява чрез корелационни коефициенти. Те измерват колко силна е връзката между тях, без да оценят количествено съотношенията между предикторите и зависимата променлива.

## Корелационен коефициент

Нека имаме две сл.в. $X$ и $Y$. Ковариацията между тях записваме с $Cov(X, Y)$ и дефинираме като:

$$Cov(X, Y) = E[XY] - E[X]E[Y]$$

Когато $Cov(X, Y)=0$ казваме, че $X$ и $Y$ са некорелирани. Позитивна или негативна ковариация измерва силата и посоката на корелацията.

Корелационният коефициент на две сл.в. $X$ и $Y$ с положителна вариация $\rho(X, Y)$ е дефиниран като:

$$\rho(X, Y) = \frac{Cov(X, Y)}{\sqrt{Var(X)Var(Y)}}$$

$\rho$ е дефиниран в интервала $[-1; 1]$.

### Видове корелационни коефициенти

Съществуват различни видове корелационни коефициенти. Според скалите, на които са измерени сл.в. между които се измерва силата на зависимостта те биват:

- параметрични
- рангови
- непараметрични

Когато сл.в. са количествени, корелационните коефициенти, които измерват зависимостта между тях се наричат **параметрични**. Те могат да бъдат единични, множествени или частно-множествени. Коефициентите са стандартизирани в интервалите $[-1; +1]$ или $[0; 1]$ и могат да се прилагат към всякакъв вид количествени променливи.

Ранговите коефициенти са предназначени за измерване на зависимостта между ординални сл.в. Съществуват два основни коефициента от тази група - коефициент на Кендал и коефициент на Спирмън. За пресмятането им се използва подредбата на единиците според значенията на променливите и съответствията между тях.

Когато имаме наблюдения представени чрез слаби статистически скали, зависимостта между тях измерваме с **непараметрични** корелационни коефициенти. Тези коефициенти са най-разнообразни, поради различните скали, които се използват. Когато имаме номинални или смесени номинални и ординални скали се прилагат предимно коефициенти, чиято формула ползва $\chi^2$ статистиката. Коефициентът на Крамер се счита за универсален, тъй като е стандартизиран в интервала $[0; 1]$ и се интерпретира като обикновен корелационен коефициент. Други използвани коефициенти са тези на асоциация и конкордация.

### Обикновени корелационни коефициенти

Обикновен коефициенти измерват зависимостта между една зависима и една предсказваща променлива, когато и двете са представени чрез силни статистически скали.

#### Коефициент на корелация на Пирсън

Коефициентът на Пирсън е един от най-прилаганите обикновени коефициенти. Записва се като:

$$R_{YX} = \sqrt{1 - \frac{S^2_Y}{\sigma^2_Y}}$$

където $S^2_Y$ е остатъчната дисперсия на зависимата променлива, а $\sigma^2_y$ е общата дисперсия. Като вземем впредвид връзката между общата, обяснената и остатъчната дисперсия, може да дефинираме коефициента като:

$$R_{YX} = \sqrt{\frac{\sum_{i=1}^n(\hat{Y}_i - \overline{Y})^2}{\sum_{i=1}^n(Y_i - \overline{Y})^2}}$$

Коефициентът е дефиниран в интервала $[0; 1]$.

#### Коефициент на Браве

Коефициентът на Браве, наричан още коефициент на линейна корелация, се пресмята като:

$$r_{YX}=\frac{\sum(X_i - \overline{X})(Y_i - \overline{Y})}{n \sigma_x \sigma_y}$$

Този коефициент измерва достоверно силата на зависимост между променливите, само когато тя е линейна. Получените ст-сти са в интервала $[-1; +1]$, т.е. освен силата на зависимост, показва и посоката.

#### Скала на интерпретация

- При $0 \leq R_{YX}  < 0.3$ корелационната зависимост се смята за слаба
- При $0.3 \leq R_{YX} < 0.5$ корелационната зависимост е умерена
- При $0.5 \leq R_{YX} < 0.7$ корелационната зависимост е значителна
- При $0.7 \leq R_{YX} < 0.9$ корелационната зависимост е силна
- При $0.9 \leq R_{YX} < 1.0$ корелационната зависимост е много силна
- При $R_{YX} = 1$ корелационната зависимост преминава във функционална
