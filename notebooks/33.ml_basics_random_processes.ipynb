# Случайни процеси

Случаен (стохастичен) процес $X_t, t \in T$, ще наричаме съвкупност от сл.в. дефинирани в общо вероятностно пространство, където $T$ е някое от множествата: $N, Z, Q, R, R^n$. Обикновено индексиращият параметър $t$ е време, но това не е задължително.

Случайните процеси се използват за моделиране на случайни експерименти, които се променят във времето:

- Време на получаване на пакети в компютърна мрежа
- Отбелязани голове на футболен отбор в поредни игри
- Дневна цена на акция в стоковата борса
- Печалби или загуби на комарджия

## Дискретни и непрекъснати сл. процеси

За сл. процес, казваме, че е дискретен, ако $T$ е крайно или изброимо множество, например:

- $N = {0, 1, \ldots}$
- $Z = {\ldots, -2, -1, 0, +1, +2, \ldots}$

Дискретните сл. процеси може да съдържат дискретни, непрекъснати и смесени сл.в. 

Непрекъснат, ще наричаме процес, за който $T$ е непрекъснато множество.

## Бял шум

Нека $T = Z$, $X_i : EX = 0$, $VarX_i < \infty$ и $X_i$ са независими сл.в. Тогава семейството ${X_i, i \in T}$ образува случаен процес, наречен **бял шум**.

Нека $Т = N$ и ${X_i, i \in T}$ образуват бял шум. Тогава $S_n = \sum_{i = 1}^n X_i$ образуват сл. процес, който се нарича случайна разходка (случайно лутане).

## Гаусов процес

Нека $T = R$ и крайномерните разпределения са многомерни нормални разпределения. Тогава сл. процес се нарича Гаусов. Този клас процеси се определят напълно от чрез математическото очакване и ковариационната матрица.

## Винеров процес

Стохастичен процес ${W_t, t \in R^+}$, за който:

- $W_o = 0$
- $W_t - W_s \in N(0, |t - s|\sigma^2)$
- $W_t$ е процес с независими нараствания
- Траекториите на процеса са непрекъснати по $t$ за всяко $w$.

се нарича Винеров процес. Винеровият процес е Гаусов.

## Брауново движение

Брауновото движение е процес, който описва движението на частиците в течност или газ.

Нека с $X_t$ означим $x$ координатата на брауновата частица в момента $t$ и нека $X_0 = 0$. Тогава ${X_i, t \geq 0}$ е стохастичен процес, като:

- Траекториите му са непрекъснати
- Процесът е еднороден, т.е. разпределението на $X_{t + h} - X_t$ е еднакво с това на $X_h$.
- $X_t$ е с независими нараствания

Брауновото движение е Винеров процес.

# Марковски вериги (МВ)

Марковстава верига наричаме дискретен сл. процес $X_n$, чиито бъдещи стойности зависят само от текущата, но не и от миналите стойности. Ако $X_n = i$ ще казваме, че процесът е в състояние $i$ в момент $n$. Когато процесът е в състояние $i$ съществува фиксирана вероятност $p_{ij}$, така че в следващия момент е в състояние $j$ и тези вероятности удовлетворяват равенствата:

$$P(X_{n + 1} = j|X_n = i, \ldots, X_1=i_1, X_0 = i_0) = p_{ij}$$

за всяко $i, j$ и всяко $n \geq 0$.

За вероятностите за преход $p_{ij}$ от състояние $i$ в $j$ е в сила:

$$p_{ij} \geq 0, \quad \sum_{j=0}^\infty p_{ij} = 1$$

тъй като процесът в момент $n + 1$ ще бъде в едно от възможните състояния.

Матрицата $P$ наричаме матрица на преходните вероятности $p_{ij}$.

## Хомогенна Марковска верига

Ако вероятността $P(X(t + s) = j| X(s) = i)$ не зависи от $s$, то МВ се нарича хомогенна, т.е. $P$ не се променя в никоя стъпка $t$.

## Свойства

### Неразложимост

За дадена МВ $X_n$ с пространство на състоянията $S = {s_1, \ldots, s_k}$ и матрица на преходните вероятности $P$ казваме, че е неразложима, ако за всички $s_i, s_j \in S$ имаме $P(s_i, s_j) > 0$.

### Поглъщащо състояние

Едно състояние се нарича поглъщащо, ако никое друго състояние не може да бъде достигнато от него.

### Периодичност

Състоянието $i$ се нарича периодично с период $d \geq 1$ тогава и само тогава, когато най-големият общ делител на числата $n$, за които $p_{ii}^n > 0$ е d, т.е. $d$ дефинира броят пъти в които веригата се връща в състояние $i$.

Състояние с период $1$ се нарича апериодично.

МВ е апериодична, ако всички състояния са апериодично. В противен случай е периодична.

## Стационарно разпределение за МВ

Стационарно разпределение за МВ е вероятностно разпределение, което остава непроменено за МВ с промяна на $t$. За дадена неразложима и апериодична МВ $X_n$, векторът $\pi = (\pi_1, \ldots ,\pi_k)$ е стационарно разпределение, ако удовлетворява:

$$\pi_i \geq 0 \> \text{за} \> i=1, \ldots ,k, \text{и} \sum_{i=1}^{k}\pi_i=1$$
$$\pi P = \pi, \sum_{i=1}^{k}\pi_i P_{i,j}=\pi_j \>\text{за}\> j=1, \ldots ,k$$

В практиката стационарните разпределния се намират чрез т.нар **Power Method**.
