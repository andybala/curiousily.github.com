# Проверка на хипотези

Най-добрият начин за проверка на това дали дадена статистическа хипотеза е вярна би бил разглеждането на генералната съвкупност. Поради непрактичността за извършване на подобни наблюдения, често се използват извадки. Ако наблюдаваните данни не са консистентни с хипотезата, тя есе отхвърля.

## Кога и къде се използват?

## Нулева и алтернативна хипотеза

**Нулевата хипотеза** предполага, че наблюдаваните данни са създадени от случаен процес. Означава се с $H_0$.

**Алтернативната хипотеза** се опитва да покаже, че има неслучаен процес, който влияе на създаването на данните. Означава се с $H_1$.

Предполагаме, че $H_0$ е вярна. Крайното решение се взима на базата на достатъчно основание за взимането му. Ако имаме достатъчно основание отхвърляме $H_0$. Ако нямаме - не я отхвърляме.

## Видове грешки

**Грешка от I род** се нарича отхвърляне на правилна $H_0$ хипотеза. Вероятността да се допусне такава грешка се нарича **ниво на значимост** и се означава с $\alpha$

**Грешка от II род** се нарича неотхвърлянето на $H_0$, когато тя е неправилна. Вероятността да се допусне грешка от втори род се означава с $\beta$

## Стъпки при проверяване на хипотези

1. Определяне на $H_0$ и $H_1$
2. Фиксира се стойноста на нивото на значимост $\alpha$. Обикновено тя се избира от някои от следните стойности - 0.05, 0.01 и 0.001
3. Избира се подходяща статистика $\mathbb{K}=K(X_1, \ldots, X_n)$, която има предварително известно разпределение, при предположение, че $H_0$ е вярна. Случайната величина $\mathbb{K}$, чрез която се проверява хипотеза се нарича **критерий**. Формулата $\mathbb{K}=K(X_1, \ldots, X_n)$ чрез която критерия се изразява посредством наблюдаваните стойности се нарича **статистика на критерия** $\mathbb{K}$. Изчислената стойност на статистиката се означава с $\mathbb{K}_{набл.}$ се нарича **наблюдавана стойност на критерия**.
4. Определя се критичната област в зависимост от нивото на значимост $\alpha$ и алтернативната хипотеза $H_1$. Множеството от всички възможни стойности на $\mathbb{K}$ на две непресичащи се множества $W$ и $\overline{W}$. Множеството $W$ се нарича **критична област** и се състой от всички стойности на $\mathbb{K}$, при които $H_0$ се отхвърля. $\overline{W}$ се нарича **област на допустими стойности** и се състой от всички стойности на $\mathbb{K}$, при които основната хипотеза се приема.

### Критични точки

Тъй като $\mathbb{K}$ е едномерна сл.в., нейните възможни стойности принадлежат на даден интервал. Следователно $W$ и $\overline{W}$ също са интервали и съществуват точки, които ги разделят.

**Критични точки** наричаме точките, отделящи критичната област $W$ и областта на допустими стойности $\overline{W}$. Те биват:

- **едностранна дясна** $[c_r, +\infty)$, която се задава с неравенството $\mathbb{K} \geq c_r > 0$;
- **едностранна лява** $(-\infty, c_l]$, която се задава с неравенството $\mathbb{K} \leq c_l < 0$;
- **двустранна** $(-\infty, c_1]\cup[c_2, +\infty)$, която се задава с неравенствата $\mathbb{K} \leq c_1$ и $\mathbb{K} \geq c_2$
