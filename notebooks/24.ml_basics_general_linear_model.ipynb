# Общ линеен модел (General Linear Model)

Общият линеен модел (ОЛМ) е съвкупност от множество модели като: t-test, analysis of variance (ANOVA), проста линейна регресия и логистична регресия. ОЛМ е обобщение на множествената линейна регресия за повече от една зависима променлива.

Нека $y$ е матрица от зависими променливи, $X$ е матрица с независими променливи, $b$ е матрица с параметри, които искаме да оценим, а $e$ е матрица съдържаща грешките или т.нар. шум. Тогава, ОЛМ може да се запише като:

$$y = Xb + e$$

Допускаме, че грешките са независими между измерванията и следват многомерно нормално разпределение. Ако това не е така може да бъде използван обобщен линеен модел (generalized linear model) за релаксиране (relaxation) на предположението относно $Y$ и $U$.

## Множествена линейна регресия

Множествената линейна регресия е обобщение на простата линейна регресия, като се вземат под предвид повече от една независима променлива. Нека отново имаме разглежданият модел:

$$y = Xb + e$$

**Теорема** Ако $X$ има пълен ранг $m$, оценката на неизвестните параметри $b$ по метода на най-малките квадрати е

$$\hat{e} = (X'X)^{-1}X'y$$
$$\text{cov}(\hat{e}) = \sigma^2(X'X)^{-1}$$

Оценката $\hat{e}$ е неизместена, ефективна и съвпада с оценката по метода на максимално правдоподобие.

# Приложения

Едно приложение на ОЛМ е анализиране на множество мозъчни рентгенови снимки в научни експерименти, където $Y$ съдържа данни от снимките, а $X$ редица от променливи, избрани от учените.
