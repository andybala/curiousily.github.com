# Математическа статистика

Основна цел на математическата статистика е изграждането на модели с помощта на теория на вероятностите. Тя дава средства за тяхната проверка и интерпретация на получените резултати.

Математическата статистика разглежда следния вероятностен модел. За всяко наблюдение, предполагаме, че е сл. в. Ако наблюденията са много, обикновено се предполага, че те са независими. Когато независимостта е съмнителна, предполагаме някакво съвместно разпределение на тези сл.в. Въз основа на това се изследват вероятностните свойства на различни функции от наблюденията - разпределения, моменти и други.

Когато за тези функции използваме реалните наблюдения, като аргументи, получаваме конкретни стойности. Въз основа на тях се правят заключения - статистически изводи, за самия модел. Идеята на тези изводи е, че при верен модел знаем доколко те са "вероятни". Така се опитваме да разберем какви са параметрите и до колко той е правдоподобен. Ако моделът се окаже неадекватен - строим друг.

## Генерална съвкупност и извадка

**Генерална съвкупност (ГС)** наричаме множеството от обекти на изследване, т.е. когато изследваме дадена х-ка на индивиди в обща съвкупност, може да отъждествим тази съвкупност със съвкупността от възможните стойности на измерваната х-ка.

В практиката често е трудно или невъзможно да се работи с всички индивиди в дадена съвкупност. Поради това работим с **извадка** от ГС, за да оценим интересуващите ни характеристики на ГС. Отделните ст-ти на х-ката в реда на получените данни наричаме **наблюдения**. Техният брой наричаме **обем на извадката**.

Основна цел при подбор на извадка с минимален обем да се направи правдоподобно заключение за ГС. Извадките се определят като систематични, случайни или подходящи за целите на изследването, в зависимост от начина по който са съставени.

Когато дадена извадка съдържа в себе си всички особености на ГС (за да представи най-добре нейната цялост) я наричаме **репрезентативна**.

При репрезентативна извадка е удобно да приемем, че тя представлява съвкупност от независими реализации на сл.в. $X$, която представя измерваната х-ка на ГС. В такива случаи последвателните набл. ще могат да бъдат представени с получената редица от измервания $x_1, \ldots, x_n$.

Векторът $\overrightarrow{x} = (x_1, \ldots, x_n)$ се третира като единично наблюдение в сложен опит, състоящ се от последователното провеждане на $n$-те наблюдения. Съвкупността $\Xi$ от възм. ст-ти на $\overrightarrow{x}$ се нарича извадково пространство. Неговата плътност на разпределение се нарича **функция на правдоподобие** (ФП) и обикновено бележим с $L(\overrightarrow{x}) = L(x_1, \ldots, x_n)$.

Ако $X$ има вероятностна плътност на разпределение $f(x)$ или дискретно разпределение ${p_x}$, ФП ще се представя с израза:

$$L(\overrightarrow{x}) = \prod_{i=1}^nf(x_i) \quad \text{или} \quad L(\overrightarrow{x}) = \prod_{i=1}^np_{x_i}$$

## Задачи на Математическата статистика

### Задачи, свързани с оценка на вероятността

Нека е дадено събитие $A$ с вероятност $p$, чиято стойност не знаем. По резултати от проведените опити трябва да оценим числото $p$, тъй като априорните съждения не са достатъчни. Тук, ГС е съвкупност от всички възможни опити, при които може да се осъществи $A$. Извадката са резултатите от проведените опити, които на практика са извършени.

### Задачи, свързани с оценка на неизвестни функции на разпределение

Тези задачи включват оценка на неизвестна вероятност. Ако дефинираме сл.в. $X=1$, когато се осъществи $A$, и $X=0$, когато не се осъществи $A$, определянето на $P(A)$ е частен сл. от задачата за определяне разпределението на $X$. В много случаи не знаем нищо за интересуващата ни функция на разпределение $F(x)$, освен, че дефиницията й като вероятност $P{X < x} = F(x)$.

### Задачи за оценка на неизвестни параметри на наблюдаваната сл.в. $X$

Тези задачи включват оценката както на параметри, влизащи в явния вид на функцията на разпределение (напр. $\alpha$ и $\sigma^2$ в нормалното разпр.), така и на други параметри, свойствени за самата величина. Оценките могат да бъдат точкови (конкретно число) или интервални (интервал, в който с известна вероятност се твърди, че се намира истинската му стойност).

### Задачи за проверка на хипотези

Широк клас от задачи в които по някакви съображения може да се изгради хипотеза, че търсена функция на разпределение е $F(x)$. Как да проверим, че хипотезата е вярна? Към този клас задачи се отнасят и проверка дали наблюдаваните две случайни величини имат еднакви средни ст-ти, даи разпределението на изследваната величина е от предполагаем клас и други.

Задачите от този тип за проверка на хипотези се отнасят към различни раздели на мат. статистика - параметрична и непараметрична статистика, многомерен анализ и др.

### Задачи за установяване на статистически зависимости между събития и величини

Това е обширен клас задачи, включващ проверки за независимост на наблюденията в извадката, установяване степента на зависимост на две сл.в., установяване на мат. вид на количествената зависимост между две съвкупности от сл.в. Про това може да е известна функ. форма, но да са неизв. някои коеф. в нея. Когато даден основен показател зависи от голям брой сл.в. е интересно те да се класифицират по степента на въздействиетото им върху този показател и да се знае кои от величините влияят благоприятно върху него и кои не. Този клас задачи се решават от множество методи - корелационен, регресионен, дисперсионен, факторен, дискриминативен, компонентен и други анализи.

# Достатъчни статистики

**Дефиниция** Статистиката $t_n(\overrightarrow{x})$ се нарича достатъчна за семейството $F$, ако условната вероятност

$$P_{\theta}\{\overrightarrow{x} \in A|t_n(\overrightarrow{x}) = t\} = P\{\overrightarrow{x} \in A\}$$

не зависи от параметъра $\theta$

**Дефиниция** Статистиката $t_n(\overrightarrow{x}),\, \overrightarrow{x}=(x_1, \ldots, x_n)$ се нарича достатъчна за семейството $F = \{F(x,\theta), \theta \in \Theta\}$, ако

$$L(\overrightarrow{x}\,|\,\theta)=h(t_n, \theta)g(\overrightarrow{x})$$

където $g(\overrightarrow{x})$ не зависи от параметъра $\theta$.

**Теорема за факторизацията** Дефиниция 1 и 2 на достатъчна статистика са еквивалентни.
